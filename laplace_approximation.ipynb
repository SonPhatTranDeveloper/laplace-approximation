# Laplace Approximation for Gamma density

```python
import numpy as np
import scipy
import matplotlib.pyplot as plt
```

## Exploring the standardised Gamma distribution

```python
# Create a Gamma distribution
a = 10
gamma_dist = scipy.stats.gamma(a)

# Plot the Gamma distribution pdf
# Defining the range
x = np.linspace(gamma_dist.ppf(0), gamma_dist.ppf(0.99), 100)

# Plot the PDF
plt.plot(x, gamma_dist.pdf(x))
plt.title("Standardised Gamma Distribution")
plt.show()
```

## Laplace approximation

```python
# Find the mode of the standardised distribution
z_0 = a - 1

# Find the variance
v = (a - 1) / (z_0 ** 2)

# Create Normal distribution
scale = np.sqrt(1 / v)
loc = z_0

# Draw the Normal distribution
normal_distribution = scipy.stats.norm(loc=loc, scale=scale)

# Define the range
range = np.linspace(normal_distribution.ppf(0.003), 
                    normal_distribution.ppf(0.997), 
                    100)

# Plot normal distribution and gamma distribution
plt.plot(range, normal_distribution.pdf(range), color='blue')
plt.plot(x, gamma_dist.pdf(x), color='orange')
plt.title("Approximating Gamma distribution with Laplace approximation")
plt.savefig("figures/fig.jpg")
plt.show()
```
